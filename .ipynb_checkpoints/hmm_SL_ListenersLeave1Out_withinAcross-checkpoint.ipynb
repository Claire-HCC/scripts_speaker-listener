```python
# this takes around 5hr for one roi
import brainiak.eventseg.event
import numpy as np
import glob
import os
import scipy.io
from scipy.stats import stats
import logging
import matplotlib.pyplot as plt
from scipy.stats import norm, zscore, pearsonr, ttest_1samp
import pickle
import time

# logging.basicConfig(level=logging.DEBUG)
```

```python
expdir = '/scratch/claire/speaker-listener/'
timeUnit='tr'
froidir='mor';
exps=['pieman','bronx','merlin','sherlock']
exp='merlin'
rname='vPCUN'
# load eventN_test
eventN_test=range(10,121)
nPerm=1000
w=5
eventN_test=range(10,121)
```

```python
d=np.load(expdir +  exp+ '/fmri/hmm/'+ rname + '_ListenersLeave1Out_eventLabels.npz')
eventLabels_self=d['eventLabels_self']
eventLabels_self.shape
```

```python
start_time = time.time()

# load mat file
for exp in [exps[2]]:
    for fname in  glob.glob(os.path.join(expdir + '/' + exp + '/fmri/timeseries/' + timeUnit + '/roi/' + froidir , 'zscore_listenerAll_'+rname+'.mat')):
        data_mat=scipy.io.loadmat(fname)
        gdata=data_mat['gdata']
        
        tn=gdata.shape[1]
        subjn=gdata.shape[2]
        voxn=gdata.shape[0]
        
        within_across_null = np.empty([np.max(eventN_test),subjn, nPerm+1])
        within_across_null[:]=np.nan
        within_across_real = np.empty([np.max(eventN_test),subjn])    
        within_across_real[:]=np.nan
        for s in range(subjn):
            if np.sum(~np.isnan(gdata[:,:,s]))>0:
                self=gdata[:,:,s]

                corrs = np.zeros(tn-w)
                for t in range(tn-w):
                    corrs[t] = pearsonr(self[:,t],self[:,t+w])[0]
                
                for K in eventN_test:
                    Ki=K-1
                    eventLabels=eventLabels_self[Ki,:,s]
                    _, event_lengths = np.unique(eventLabels, return_counts=True)
                    
                    within = corrs[eventLabels[:-w] == eventLabels[w:]].mean()
                    across = corrs[eventLabels[:-w] != eventLabels[w:]].mean()
                    within_across_real[Ki,s] = within - across

                    # Compute within vs across boundary correlations, for real and permuted bounds
                    np.random.seed(0)
                    for p in range(nPerm+1):
                        perm_lengths = np.random.permutation(event_lengths)
                        eventLabels = np.zeros(tn, dtype=np.int)
                        eventLabels[np.cumsum(perm_lengths[:-1])] = 1
                        eventLabels = np.cumsum(eventLabels)

                        within = corrs[eventLabels[:-w] == eventLabels[w:]].mean()
                        across = corrs[eventLabels[:-w] != eventLabels[w:]].mean()
                        within_across_null[Ki,s,p] = within - across


        np.savez(expdir +  exp+ '/fmri/hmm/'+ rname + '_ListenersLeave1Out_withinAcross.npz', within_across_null=within_across_null,within_across_real=within_across_real)      


print(time.time()-start_time)    
```

```python
plt.plot(np.nanmean(within_across_real,axis=1))
plt.plot(np.nanmean(within_across_null,axis=1),'gray')
plt.show()
```

```python
realzm=np.mean(np.arctanh(within_across_real),axis=1)
nullzm=np.mean(np.arctanh(within_across_null),axis=1).T

t,_=ttest_1samp(nullzm,realzm)
```
